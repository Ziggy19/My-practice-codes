```python
# Write a Python Program to Develop PhoneBook
phonebook={'ziggy':8719912636,'muskan':6264180525,'monu':8735024484,'mizi':9558828611,'Lado':897545434}
while True:
    print("\t \t \t \t Phone Book\t")
    print("Enter 1 to Add new contact\n")
    print("Enter 2 to Update Phone Book\n")
    print("Enter 3 to Remove contact\n")
    print("Enter 4 to search Phone Book\n")
    print("Enyer 5 to Display Phone Book\n")
    print("6 for Exit\n")
    option=int(input("What do u want?"))
    if option==1:
        name=input("Type Name:\t")
        number=int(input("Type Number:\t"))
        phonebook[name]=number
        print("contact succesfully Added")
    elif option==2:
        name=input("Enter name:\t")
        if name in phonebook:
            number=int(input("Type Number:\t"))
            phonebook[name]=number   
            print("contact succsesfully updated")
        else:
            print("name not exist please add new contact")
    elif option==3:
        name=input("Enter name:\t")
        if name in phonebook:
            del phonebook[name]  
            print("contact succsesfully deleted")
        else:
            print("name not exist please add new contact")    
    elif option==4:
        name=input("Enter name:\t")
        if name in phonebook:
            print(f"{name} : \t",phonebook[name])
        else:
            print("name not exist please add new contact") 
    elif option==5:
        print(phonebook)
    elif option==6:
        print("Thank you")
        break





```
